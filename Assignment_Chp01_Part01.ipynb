**28.Define a function percent(word, text) to calculate the frequency at which a given word appears in the text.
is expressed as a percentage.**

```python
def percent(word,text):
	count = 0
	for w in text:
		if word == w:
			count+=1
	return str(float(count) / float(len(text)) *100)+'%'

a=percent("All",text6)
print(a)
```
